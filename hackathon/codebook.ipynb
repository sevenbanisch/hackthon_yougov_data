# Loading and parsing the codebook


In this notebook we load the .xlsx codebook file stored in the datapath of our repository (CONVENTION). 



```python
import pandas as pd

datapath = '../data/'
filename = 'codebook_sb'
ext = '.xlsx'
```

```python

df = pd.read_excel(datapath+filename+ext)
```

```python
#print(df)
```

```python
#df.head(10)
```

## Here we extract the IDs and texts of all questions


```python
item_dict = {}
idx = df.index[df['col3'] == 'Value'].tolist()

#print(ixs)

for ix in idx:
    #print(df.iloc[ix-1,0],df.iloc[ix+1,2])
    item_dict.update( { df.iloc[ix-1,0] : df.iloc[ix+1,2] } )
```

```python
print('\nNote that I still have not checked what is actually missing.')
print('\nBut there are',len(item_dict),'different questions in.')
```

## So let's print them:

```python
for ele in item_dict:
    print(ele,' : ',item_dict[ele])
```
