```python
filename = 'codebook_sb'
ext = '.xlsx'
df_codebook = pd.read_excel(datapath+filename+ext)

item_dict = {}
idx = df_codebook.index[df_codebook['col3'] == 'Value'].tolist()
for ix in idx:
    item_dict.update( { df_codebook.iloc[ix-1,0].lower() : df_codebook.iloc[ix+1,2] } )
```

```python
means = []
stds = []
mins = []
maxs = []
cols = []
quests = []
exclude = ['recordno','qweek','case_id']
for px,ele in enumerate(df_list):
    meansP = []
    stdsP = []
    minsP = []
    maxsP = []
    for col in ele:
        if col not in keep and col not in exclude:
            meansele = ele[col].mean()
            meansP.append(meansele)
            
            stdsele = ele[col].std()
            stdsP.append(stdsele)
            
            minsele = ele[col].min()
            minsP.append(minsele)
            
            maxsele = ele[col].max()
            maxsP.append(maxsele)
            if px < 1:
                cols.append(col)
                quests.append(item_dict[col])
    means.append(meansP)
    stds.append(stdsP)
    mins.append(minsP)
    maxs.append(maxsP)
    

#print(df_P1.mean())
#print(len(means[4]))

for ex,ele in enumerate(cols):
    print(ele,': ',mins[0][ex],maxs[0][ex],quests[ex][:80])
```
